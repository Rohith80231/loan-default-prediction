```python
df.drop('LoanID',axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['HasMortgage'] = le.fit_transform(df['HasMortgage'])
df['HasDependents'] = le.fit_transform(df['HasDependents'])
df['HasCoSigner'] = le.fit_transform(df['HasCoSigner'])
```

```python
df = pd.get_dummies(df, columns=['Education', 'EmploymentType', 'MaritalStatus', 'LoanPurpose'], dtype=int)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X =df.drop('Default',axis=1)
y = df['Default']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
y_train.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
print(y_train_smote.value_counts())
```

```python
from xgboost import XGBClassifier

model = XGBClassifier(
    scale_pos_weight=7.5,
    random_state=42,
    n_estimators=100,
    learning_rate=0.1,
    max_depth=5
)
```

```python
model.fit(X_train_smote, y_train_smote)
```

```python
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix

y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)[:, 1]

print(classification_report(y_test, y_pred))
print("ROC-AUC Score:", roc_auc_score(y_test, y_pred_proba))
```

```python
print(confusion_matrix(y_test, y_pred))
```
